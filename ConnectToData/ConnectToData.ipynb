# Working with MySQL

```python
# import mysql.connector as connection
from sqlalchemy import create_engine
import os
import pandas as pd

from dotenv import load_dotenv

load_dotenv()

conn_str = f"mysql+pymysql://{os.getenv('USER')}:{os.getenv('PASSWD')}@{os.getenv('HOST')}:{os.getenv('PORT')}/{os.getenv('DATABASE')}"    
engine = create_engine(conn_str)
query = 'SELECT * from ClimateObservations LIMIT 50'
df = pd.read_sql(query, con=engine)
df.head()
```
